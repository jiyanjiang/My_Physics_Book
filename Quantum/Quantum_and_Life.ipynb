## 1.什么是量子

量子（Quanta）：一份的意思；引申：计量能量的最小单位，能量不可再分的最小单位；

比如：对光来说有光的量子；

光是电磁波，可见光其实是不同波长的光的“混合”。

<img src="ray-prism.png" width="300">

光速是恒定的，在真空中是$c$，

$$ c= 3.0 \times 10^8 m/s $$

$$ c= \frac{\lambda}{T} = \lambda \nu $$

这里$ \lambda $是波长，$ T $是周期，$ \nu$是频率。

频率为$ \nu $光的最小能量是$h \nu$。

$$ E = h \nu $$

换句话说能量就是频率，$ h$是换算因子，被称为普朗克常数，

$$ h = 6.626 \times 10^{-34} J \cdot s = 4.136 \times 10^{-15} eV \cdot s$$

这是一个很小很小的数。

对绿光（波长500nm）而言，我们可计算出一个光量子的能量是约2.48eV

$$ E = h \nu = \frac{h c}{\lambda} $$

这意味着：频率为$ \nu $的光，它的能量必须是$ h \nu $的整数倍，即：

必须是$ h \nu$，$ 2 h \nu$，…… $ n h \nu$，……

这可以被形象的说为，1个光子，2个光子，……

光量子只是量子的一个例子——能量以一份一份的形式存在，而非连续地存在——或能量是量子化的，在自然界中是普遍存在的。

比如：氢原子的能级也是量子化的。

## 2.氢原子

<img src="https://web.njit.edu/~gary/321/H_energy_level.gif" width="350">

氢原子的物理性质，很大程度上是由氢原子中电子的运动状态决定的。

这里$ n = 1, 2, 3, ... $

电子带负电，质子带正电，正电和负电之间存在吸引力。

在“地球-卫星”模型中，卫星质量远小于地球质量，卫星和地球因“万有引力”而存在吸引力，最终导致卫星围绕地球“转圈”。

### 3.1.粒子

传统上我们认为电子是粒子。

在我们的日常经验中，粒子的行为就好像是一个“石子”的行为（或者你也可以说是一个人的行为）。

我们应如何描述一个石子呢？

它具有位置，可以用笛卡尔坐标系中的一个坐标来表示$\vec r = (x, y, z )$。但一般我们会说石子是有大小的，但这个大小是确定的，石子不可能弥散于整个空间，它一定是局限在空间中的某个确定范围内的。

比如：在我们的经验中如果一个石子在北京，那么它就不能同时在上海。

石子的运动状态还可以用速度来表示，

$\vec v = \frac{d \vec r}{d t }$

我们认为在任意时刻，一个石子都会有个位置，同时它还会有个速度。随着时间的演化，石子会动起来，划出一条“抛物线”。

如果你不理解上面的话，就从地上拾起一个石子然后抛出去，我们的眼睛可以观察到石子会划出一条轨迹，落在远处的地面上，石子的轨迹不会在空间中的某一点分叉。

当然我们还可以继续描述这个石子，比如石子的动能是

$$T = \frac{m v^2}{2}$$

石子的势能是

$$ V = mgh $$

这里$m$是石子的质量，$v$是石子的速度，$h$是石子的高度。石子的总能量是：

$$ E = T+V $$

如果我们定义石子的动量$p$是$mv$的话，石子的总能量可以改写为：

$$ E = \frac{p^2}{2m } + V $$

换句话说石子的运动状态是由石子的动量$p$和石子的位置$x$（高度$ h$是石子位置$x $的一个分量）描述的。

德布罗意认为：电子的行为可以用**波动的语言**予以描述。

### 3.3.物质波

在传统观念下，电子的运动可以用动量$p$来描述，此时电子的能量是$\frac{p^2}{2m}$。

德布罗意认为电子其实是一种波动，必须用波动的语言——波长$\lambda$来描述。

他认为在粒子语言和波动语言之间存在着一种关系（相当于是某种翻译机制），

$$ \lambda = \frac{h}{p} $$

左侧是波动的语言（出现了$\lambda$），右侧是粒子的语言（出现了$p$）。

如果电子果真是波的话，氢原子中的电子会围绕质子运动，当这个电子围绕质子运动了一圈之后，我们认为电子的状态会回到一圈前的状态，而现在电子的运动状态是用波函数来描述的。

<img src="http://i.stack.imgur.com/V7SLt.png" width="400">

电子围绕质子一圈正好等于波长的整数倍

$$ 2 \pi r = n \lambda $$

这意味着

$$ \lambda = \frac{h}{p} = \frac{2 \pi r}{n} $$

得到：

$$ r\cdot p = n\frac{h}{2 \pi} = n \hbar$$

$ r\cdot p $是轨道角动量$L_n$，上式是轨道角动量量子化条件。由这个条件可以推出氢原子能量的量子化公式$ E_n = - \frac{13.6}{n^2} eV$。

## 4.波的干涉

两列波相遇，互相干涉，会发生干涉效应。

<img src="http://boomeria.org/physicslectures/waves/interference.jpg" width="300">

干涉的结果是某些地方的波动可能会增强，某些地方的波可能会减弱。

如果是水波的干涉的话，我们可以观察水波的振幅来研究波的干涉。如果是光波的话，我们可以观察光强。

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/86/Ebohr1_IP.svg/440px-Ebohr1_IP.svg.png" width="300">

上图是光的“杨氏双缝干涉”，我们可以在屏（F）上观察到光强的明暗分布，即明亮条纹和黑暗条纹的交替分布。

作为波动性的最突出的特征是：在双缝中心正对的屏幕区域竟然是“明亮”的，在我们的直觉中，双缝正对区域本来是离光线最远的区域（假设光像粒子一样沿直线传播的话），这个区域应当是最难被光照亮的区域。

但如果我们考虑光是波动的话，双缝$b$和$c$，每一个都是新的光源，波动以$b $和$ c$为中心向四面八方传播，考虑照射到双缝正对屏幕的位置，波动传播到这个位置正好是同相位的（即两列波同步地达到波峰和波谷），这个时候波动是增强的，因此对应明亮的条纹。

相邻明亮（或黑暗）条纹的角度是：

$$ \theta = \frac{\lambda}{d} $$

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/d/dc/Double_slit.svg/600px-Double_slit.svg.png" width="300">

因为此时，通过双缝的这两列波，它们传播过的路程正好相差一个波长$\lambda$，相差一个波长意味着这两列波依然是同相位的，因此依旧是明亮的条纹。

## 5.相干时间和相干长度

像这样能够发生干涉效应，并表现为总光强明暗交替的两列波就称为**相干的**。

相干条件是：两列波振动频率相同，两列波的位相差要保持恒定。

如果两列波的振动频率稍有不同（$\omega_1 \neq \omega_2$），这两列波在有限时间（退相干时间$\tau$）内还能发生干涉现象，时间久了干涉现象就没有了。

$$ \tau = \frac{\pi}{\omega_1 - \omega_2} $$

实际的光不会只包括两个频率，而是一个频率的分布，假设分布的带宽是$\Delta \omega$，相干时间是：

$$ \tau = \frac{2 \pi}{\Delta \omega} $$

由此，还可以定义光的相干长度：

$$ l = c \tau $$

作为例子：白炽灯的带宽是$10^8 Hz$，相干时间是$\sim 6 \times 10^{-8} s$，相干长度是$\sim 18 m$；典型激光器输出的光，带宽是$10^4 Hz$，相干时间是$\sim 6 \times 10^{-4}s $，相干长度为 $\sim 180 km $。

## 7.波函数的坍缩

### 几率解释

波函数$\psi(x)$并不直接对应物理量，但波函数的绝对值的平方$\left| \psi(x) \right|^2$对应我们在位置$x $附近发现粒子的概率。

假设波函数是：

$$ \psi(x) = A e^{i \left( \frac{2 \pi  }{\lambda} \right)  x } $$

我们发现我们在空间中所有位置发现粒子的几率都是相同的。

换句话说粒子可以以一定几率在北京，同时以一定几率在上海。

在我们做测量前，我们无法百分百确认粒子在北京或在上海。

为了百分百确认粒子到底在什么地方，我们必须对粒子的位置进行测量。

这里我们先不讨论测量的物理机制。

但我们知道成功的测量必然会导致波函数的突然改变，假如我们测量的结果是粒子处于北京，那么波函数会瞬间改变为一个仅仅局域于北京附近的一个波函数。

这种波函数的突然改变就叫“波函数的坍缩”。

<img src="https://phys.libretexts.org/@api/deki/files/19905/clipboard_ef02f6d4e6fd04951f287b2a4e575409e.png?revision=1" width="300">


这将会导致一些有趣的讨论：

1. 首先制备一个波函数$\psi(x)$，这个波函数不局限在某个区域，即我们有大致相同的几率在整个空间找到这个粒子；

2. 在时刻$t_1$对波函数$\psi(x)$进行位置测量，波函数$\psi(x) $会随机地坍缩在某个位置$x_1$；

3. 改变这个波函数，使之重新成为扩展在整个空间的波函数$\psi'(x)$；

4. 在时刻$t_2$对波函数$\psi'(x)$进行位置测量，波函数$\psi'(x)$将随机坍缩到某个位置$x_2$；

由于$x_1$和$x_2$是完全随机的，这意味着

$$ \frac{x_1 - x_2}{t_1 - t_2} $$

可以超过光速$ c$。

这个结果表面上与狭义相对论的结果冲突，但考虑到以上物理机制无法对信息进行编码，传递信息，所以这个结果与狭义相对论并不冲突。

## 9.多电子原子

氢原子是最简单的原子，原子核外面只有一个电子。

如果原子核外面有很多个电子的话，电子会按照“壳层”结构一层一层地向外填充。

离原子核越近的壳层，对应的能量越低。同时能够填充的电子的数目也越小。

n=1时，最多能填充两个电子，这两个电子自旋相反。

n=2时，最多能填充8个电子。

……

对一个原子来说，如果它本身所具有的电子数目正好使得所有壳层都填满了电子，那么它的结构就是稳定的。

如果正好比满壳层多出一个电子，它就倾向于失去这个电子，使得剩余的（阳）离子部分正好是满壳层的。

如果正好缺一个电子就是满壳层，它就倾向于（从其他原子那里）获得一个电子变得是满壳层的。

当不同种类的原子互相靠近的时候，外层的电子会重新调整、配对，形成更复杂的新的稳定结构——分子。

### 氢分子

最简单的分子是氢分子，

<img src="http://astronomy.swin.edu.au/cms/imagedb/albums/userpics/molecularhydrogen1.jpg" width="300">

直观地理解，电子带负电，居于两个氢原子核（质子，带正电）的中间，按静电相互作用（异性相吸，同性相斥），这个结构就是稳定的。

实际的计算，我们需要考虑量子力学。

首先因为电子质量远远小于质子质量，我们主要考虑电子的运动，并把电子看作是量子力学的对象。

根据量子力学电子是费米子，对于费米子来说，波函数应当是“奇”的，即：交换两个电子，总波函数应该有一个负号。

$$ P_{12}\Psi(12) = - \Psi(12) $$

考虑到电子还有自旋，电子的总波函数要写成轨道部分$\psi$和自旋部分$\chi$的乘积的形式。

$$ \Psi = \psi \times \chi $$

如果要求总波函数$\Psi$是“奇”的话，$\Psi_A$，那么就有两种可能性，即或者轨道部分波函数是“奇”的，或者自旋部分波函数是“奇”的。

$$ \Psi_A = \psi_A \times \chi_S $$

$$ \Psi_A = \psi_S \times \chi_A $$

这里A表示“奇”的，或交换反对称的；S表示“偶”的，或交换对称的。

自旋部分交换反对称的波函数$\chi_A$可表示为，

$$ \chi_A = \frac{1}{\sqrt 2} \left( \left| \uparrow \downarrow \right\rangle - \left| \downarrow \uparrow  \right\rangle \right) $$

$$ \chi_S = \frac{1}{\sqrt 2} \left( \left| \uparrow \downarrow \right\rangle - \left| \downarrow \uparrow  \right\rangle \right) $$

## 10.量子纠缠

“纠缠”的字面意思是：一个东西和另一个东西融合为一个整体，我们无法指着一个东西说它就是这样的。

比如氢分子中的两个电子，它们自旋相反，构成一个化学键。

我们只写出它们自旋部分的波函数：

$$ \chi_A = \frac{1}{\sqrt 2} \left( \left| \uparrow \downarrow \right\rangle - \left| \downarrow \uparrow  \right\rangle \right) $$

对这个波函数，我们无法说出第一个电子的自旋到底是向上的，还是向下的，

我们可以说这两个电子肯定有一个自旋是向上的，另一个自旋是向下的，就是不知道哪一个是向上的还是向下的。

对氢分子中的这两个电子而言，它们离的是很近很近的，我们可以设想逐渐让这两个电子分离，一个电子向左跑，另一个电子向右跑，但我们并不改变它们的自旋部分的波函数。

当这两个电子分的足够开之后，我们再在左边对第一个电子的自旋进行测量。如果测量成功的话，它将或者是自旋向上的，或者是自旋向下的。

如果我们观察一下波函数$\chi_A$的话，我们发现我们没法确切地推测出第一个自旋的取向，因为$\chi_A$本身是两项的叠加，在这两项中，第一项$\left| \uparrow \downarrow \right\rangle$里左侧电子的自旋向上，而在第二项中$\left| \downarrow \uparrow  \right\rangle$左侧电子的自旋向下。

一次成功的测量，会使波函数$\chi_A$瞬间坍缩到两项中的任何一项，这个过程也叫“退相干”，因为本来波函数是两项的相干地叠加，但现在只能按各50%的几率处在第一项或第二项。

如果波函数是$\left| \uparrow \downarrow \right\rangle$或者$\left| \downarrow \uparrow  \right\rangle$的形式，它们就不再是“纠缠态”了，因为对$\left| \uparrow \downarrow \right\rangle$或者$\left| \downarrow \uparrow  \right\rangle$我们都可以确定地说出左侧电子自旋的取向。

![量子纠缠](https://www.picoquant.com/images/uploads/application_images/8005/quantum_entanglement.jpg)

假如我们在左侧对第一个电子做了自旋测量，其结果将是随机性的自旋向上或自旋向下，一旦完成对第一个电子的测量，波函数就已经发生了坍缩（波函数瞬间突变）。此时我们就可确定性地预言右侧第二个电子的自旋取向，它将一定与第一个电子的自旋取向相反。

量子纠缠是最近经常被人讨论的话题。比如有人做这样的比喻：

把一双手套，一只是左手的，一只是右手的，封闭在两个黑盒子里，如果不打开的话，我们当然无法确切地知道每只盒子里手套的“手性”，但我们总可以说如果一只是左手的，那么另一只就一定是右手的。

这听起来和量子纠缠很像，但其实与量子纠缠是不同的，在手套的例子里我们仅仅是不知道，是信息的缺乏，而在量子纠缠中，我们是知道的，我们并不缺乏任何信息。

## 12.光合作用

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/d/db/Photosynthesis.gif/440px-Photosynthesis.gif" width="200">


$$ 6 CO_2 \uparrow + 6 H_2 O + h \nu \to C_6 H_{12} O_6 + 6 O_2 \uparrow $$

光合作用相当于是把光能$h \nu$储存在糖$C_6 H_{12} O_6$里。

光改变的是电子的运动状态，“光合作用”相当于是光把电子从能量地形图中较低的位置（左侧反应物）“推到”能量地形图中更高的亚稳态（右侧生成物）。(联想：通过抬升重物储能。)

光合作用的意义在于糖更容易在生命体中被储存和输运（比喻：搬动电池）。

### FMO复合体

FMO复合体是一种光合作用复合体，由一种被称为“绿色硫细菌”的光合微生物合成，这种细菌分布在类似黑海深处的富硫水体中。

细菌或植物可以把光能转化为化学能。量子力学在这背后发挥了关键作用。

量子态很容易被无规则的热运动破坏，温度会破坏量子行为。但生命现象恰恰发生于相对较高的温度（$\sim 300 K$）。

量子相干更容易在低温下被维持，在高温下，量子相干会被破坏。（系统和环境耦合，环境对系统的作用会破坏相干性）

在FMO复合体中，细菌叶绿素（阵列）会“采集”光，使叶绿素处在被激发的状态，然后把能量通过“叶绿素-蛋白质”网络传给反应中心。

令人惊讶的是，上述能量转换、传输的效率相当高，几乎是100%，与之比较人造太阳能电池的转换率只有70%。

相当于叶绿素总知道走哪条路径才是通向反应中心的最短路径。如果在路上耽搁太久的话，当然会降低能量的转换率。

这也可看做是一个计算问题，计算通向反应中心的最短路径，最简单的经典算法是“随机行走”，但这个效率显然太低了（时间远大于1纳秒，将导致大量能量耗散掉）。

如果考虑这个过程是量子过程的话，能量将可以叠加的形式，同时走过所有的路径（好像费曼路径积分），对应的是一个“量子行走”。最终量子态坍缩在最短路径上，对应最佳能量传输效率。

在蛋白质网络中找到通往反应中心的最短路径等价于很多计算机科学中的其他类似问题。这里的关键是找到最佳的量子计算算法，并且这些算法在常温下也可保持高效运作，对FMO复合体的研究有利于我们找到这些算法。
